# CHALLENGE 1

```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
url = 'https://api.github.com/repos/ironhack-datalabs/madrid-oct-2018/forks'
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
```

```python
response = requests.get(url,headers=headers)
results = response.json()
repos = json_normalize(results)
print(repos)
```

```python
repos.columns
```

```python
df_repos = pd.DataFrame(repos)
repos.head
```

```python
list_forks = [i for i in df_repos['forks_url']]
list_forks
```

```python
list_of_lang = []
for lang in df_repos['language']:
    if lang not in list_of_lang:
        list_of_lang.append(lang)
```

```python
print(list_of_lang)
```
